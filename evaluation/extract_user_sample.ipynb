```python
## Extract Users Evaluation Sample

#  Notebook que extrai amostra de usuários para avaliação
```

```python
import pandas as pd
```

```python
df_review = pd.read_csv("../data/yelp_dataset/yelp_academic_dataset_review.csv")
df_review.head()
```

```python
df_review.shape 
```

```python
df_review_total = df_review\
                    .groupby("user_id").agg(
                      total_reviews = ("business_id", "count")
                    ).reset_index()
df_review_total.head()
```

```python
df_review_with_5 = df_review[df_review['stars'] >= 5]\
                    .groupby("user_id").agg(
                      business_with_5 = ("business_id", list), 
                      total_with_5 = ("business_id", "count")
                    ).reset_index()
df_review_with_5.head()
```

```python
df_users = df_review_total\
            .merge(df_review_with_5, on="user_id", how="inner")

# Filter only users with 5 or more reviews
df_users = df_users[df_users['total_reviews'] >= 10].sample(1000, random_state=42)
df_users.shape
```

```python
df_users
```

```python
df_eval_review = df_review\
                .merge(df_users, on="user_id", how="inner")[df_review['stars'] < 5]\
                .groupby("user_id").agg(
                  business_less_5 = ("business_id", list), 
                  total_less_5 = ("business_id", "count")
                )\
                .merge(df_users, on="user_id", how="inner")\
                .reset_index()

df_eval_review.head()
```

```python
import numpy as np

business_ids = list(df_review.merge(df_users, on="user_id", how="inner").business_id.unique())

df_eval_review["business_random"] = df_eval_review.apply(lambda x: list(np.random.choice(business_ids, 10, replace=False)), axis=1)
df_eval_review.head()
```

```python
# join two columns lists in dataframe
import random

df_eval_review["reclist"] = df_eval_review.apply(lambda x: x.business_random + x.business_less_5, axis=1)
df_eval_review["reclist"] = df_eval_review.apply(lambda x: random.sample(x.reclist, len(x.reclist)), axis=1)
df_eval_review["total_reclist"] = df_eval_review["reclist"].apply(len)
df_eval_review.head()
```

```python
df_eval_review[["user_id", "business_with_5", "total_with_5", "business_less_5", "total_less_5", "reclist"]]\
  .to_csv("../data/evaluation/eval_users.csv", index=False)
```
